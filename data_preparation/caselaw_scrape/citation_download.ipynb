```python
from eyecite import get_citations, annotate_citations, clean_text
from eyecite.models import FullCaseCitation

def get_unique_citations(text, idx):
    cleaned_text = clean_text(text, ['all_whitespace'])
    citations = get_citations(cleaned_text)
    unique_full_case_citations = []
    section_seen_combinations = set()
    # section_seen_combinations is a set to track unique (reporter, volume, page) combinations
    
    for citation in citations:
        if isinstance(citation, FullCaseCitation):
            try:
                to_lookup = citation.groups
                reporter = to_lookup['reporter']
                volume = to_lookup['volume']
                page = to_lookup['page']

                combination = (reporter, volume, page)

                if combination not in section_seen_combinations:
                    unique_full_case_citations.append(citation)
                    section_seen_combinations.add(combination) 
                    
            except KeyError as e:
                print(f"Missing key {e} in citation: {citation}")
                print(f"The index is: {idx}\nThe citation text is: {citation.matched_text()}\n")
                continue 

            except Exception as e:
                print(f"Unexpected error: {e} in citation: {citation}")
                print(f"The index is: {idx}\nThe citation text is: {citation.matched_text()}\n")
                continue 
                
    return unique_full_case_citations
```

```python
# Even newer version that will retry finding a case if throttled
import requests
import json
import time
from bs4 import BeautifulSoup

def fetch_combined_sub_opinions(citation, api_token):
    # Extract volume, reporter, and page from the citation object
    citation_data = citation.groups
    reporter = citation_data['reporter']
    volume = citation_data['volume']
    page = citation_data['page']
    
    cite_endpoint = "https://www.courtlistener.com/api/rest/v4/citation-lookup/"
    headers = {
        'Authorization': f'Token {api_token}',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.107 Safari/537.36'
    }

    data = {
        'volume': volume,
        'reporter': reporter,
        'page': page,
    }
    
    # Dictionary to store the citation as the key and the list of sub-opinions as the value
    citation_sub_opinions = {}

    while True: 
        try:
            response = requests.post(cite_endpoint, headers=headers, data=data)
            
            if response.status_code == 200:
                main_data = response.json()
                combined_sub_opinions = []
                
                if main_data[0]['status'] in [200, 300]:
                    found_combined = False
                    for cite in main_data:
                        if found_combined:
                            break 
                        for cluster in cite['clusters']: 
                            if found_combined:
                                break 
                            for sub_opinion_url in cluster['sub_opinions']:
                                if not found_combined:
                                    sub_opinion_response = requests.get(sub_opinion_url, headers=headers)
                                else:
                                    break  # Already found the combined opinion, stop searching

                                if sub_opinion_response.status_code == 200:
                                    sub_opinion_data = sub_opinion_response.json()
                                    
                                    # Check if the sub-opinion "type" is "010combined", which will include main opinions, concurrances, and dissents.
                                    if sub_opinion_data.get('type') == "010combined":
                                        for field in fields_to_check:
                                            if field in sub_opinion_data and sub_opinion_data[field]:  # Check if field exists and is non-empty
                                                cleaned_text = BeautifulSoup(sub_opinion_data[field], "html.parser").get_text()
                                                buf = {}
                                                # buf[main_data[0]["citation"]] = cleaned_text  # I don't want to add a key here I think
                                                combined_sub_opinions.append(cleaned_text)
                                                found_combined = True
                                                break  # Break out of searching each text field
                    if found_combined:
                        citation_sub_opinions[(reporter, volume, page)] = combined_sub_opinions
                    else:
                        citation_sub_opinions[(reporter, volume, page)] = "<MISSING>"
                        # print(f'Could not find the combined opinion for {citation.matched_text()}')
                elif main_data[0]['status'] == 429:
                    print("You got a 429 status code, what does it look like?")
                    print(json.dumps(main_data))
                else:
                    citation_sub_opinions[citation.matched_text()] = "<MISSING>"
                    # print(f"Could not retrieve data for case: {main_data[0]['citation']}. Error code: {main_data[0]['status']} Error message: {main_data[0]['error_message']}")

                break  # Exit the loop once the request succeeds

            elif response.status_code == 429:
                # If a 429 status code is received, read the Retry-After header
                retry_after = int(response.headers.get('Retry-After', 60))  # Default to 60 seconds if the header is missing
                print(f"Rate limit exceeded fetching case: {citation.matched_text()}. Retrying after {retry_after} seconds.")
                time.sleep(retry_after)
            else:
                # print(f"Failed to retrieve data for citation: {citation.matched_text()}. Status: {response.status_code}, Response: {response.text}")
                citation_sub_opinions[(reporter, volume, page)] = "<MISSING>"
                break  # Exit the loop, no point in retrying on these status codes

        except Exception as e:
            print(f"Unexpected error for citation: {citation.matched_text()}. Error: {e}")
            citation_sub_opinions[(reporter, volume, page)] = "<MISSING>"
            break  # Exit the loop on an unexpected error

    return citation_sub_opinions
```

```python
from eyecite import get_citations, annotate_citations, clean_text
from eyecite.models import FullCaseCitation

def annotate_text(text):
    if text is None:  # Check for None before processing
        return None, None
    cleaned_text = clean_text(text, ['all_whitespace'])
    citations = get_citations(cleaned_text)
    full_case_citations = []
    uuid_citations = {}
    seen_combinations = set()  # Set to track unique (reporter, volume, page) combinations
    
    for citation in citations:
        if isinstance(citation, FullCaseCitation):
            try:
                to_lookup = citation.groups
                reporter = to_lookup["reporter"]
                volume = to_lookup["volume"]
                page = to_lookup["page"]
                # Add to citations if all required keys exist
                cite_key = f"{volume} {reporter} {page}"
                citation_data = {
                    "citation_obj": citation,
                    "span_start": citation.full_span_start,
                    "cite_key": cite_key
                }
                full_case_citations.append(citation_data)
                # full_case_citations.append((citation, citation.full_span_start, cite_key))
            except KeyError:
                # Skip invalid citation
                continue
                
    adjusted_text = annotate_citations(cleaned_text, [[c["citation_obj"].full_span(), "<MASK>", "</MASK>"] for c in full_case_citations], source_text=cleaned_text)
    search_start = 0

    while True:
        # Find the next <MASK> token
        mask_start = adjusted_text.find("<MASK>", search_start)
        if mask_start == -1:
            break  # No more <MASK> tokens

        mask_end = adjusted_text.find("</MASK>", mask_start)
        if mask_end == -1:
            raise ValueError("Unmatched <MASK> and </MASK> tokens.")

        united_offset = 7  # Length of "United "
        if mask_start >= united_offset and adjusted_text[mask_start - united_offset:mask_start] == "United ":
            # Remove "United " by slicing it out of the adjusted_text
            adjusted_text = adjusted_text[:mask_start - united_offset] + adjusted_text[mask_start:]
            mask_start -= united_offset  # Adjust mask_start to account for the removal of "United "
            mask_end -= united_offset

        masked_text = adjusted_text[mask_start + len("<MASK>"):mask_end]

        citation_uuid = str(uuid.uuid4())

        masked_citations = get_citations(masked_text)

        if not masked_citations:
            # Remove the <MASK> and </MASK> tokens, and the text in between
            adjusted_text = (
                adjusted_text[:mask_start] +
                adjusted_text[mask_end + len("</MASK>"):]
            )
            search_start = mask_start  # Continue searching after the removed portion
            continue

        # Store the first result in the uuid_citations dictionary as get_citations returns a list
        uuid_citations[citation_uuid] = masked_citations[0]

        # Replace the text between <MASK> and </MASK> with the UUID
        adjusted_text = (
            adjusted_text[:mask_start + len("<MASK>")] +
            citation_uuid +
            adjusted_text[mask_end:]
        )

        # Update the search start position to avoid infinite loops
        search_start = mask_start + len("<MASK>") + len(citation_uuid)
        
    return uuid_citations, adjusted_text
```

```python
def process_sections(row):
    updated_sections = []  # To hold processed sections
    citation_dict = {}     # To map UUID to citations
    annotated_dict = {}   

    for section in row["sections_alt"]:
        for key, value in section.items():
            section_uuid = str(uuid.uuid4())
            
            # Process text
            citations, annotated_text = annotate_text(value)
            ''' # using lists
            if citations is not None:
                all_citations.extend(citations)
            if annotated_text is not None:
                annotated_texts.append(annotated_text)
            '''
            
            # Add to dictionaries if valid
            if citations is not None:
                citation_dict[section_uuid] = citations
            else:
                citation_dict[section_uuid] = []
                
            if annotated_text is not None:
                annotated_dict[section_uuid] = annotated_text
            else:
                annotated_dict[section_uuid] = None
            
            # Store in dictionaries keyed by UUID
            citation_dict[section_uuid] = citations
            annotated_dict[section_uuid] = annotated_text
            
            # Create updated section with UUID
            updated_section = {
                "section_id": section_uuid,
                "section_header": key,
                "original_text": value,
                "annotated_text": annotated_text,
                "citations": citations,
            }
            updated_sections.append(updated_section)

    row["sections_final"] = updated_sections

    return row
```
